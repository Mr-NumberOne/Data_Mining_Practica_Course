### Removing Numbers & newline From `'Company Names'` 

```python
unclean_data_set['Company Name']=unclean_data_set['Company Name'].apply(lambda x:x.split("\n")[0])
# + taking the company name and ignoring anything not in the first line
```

```python
unclean_data_set['Company Name']=unclean_data_set['Company Name'].str.replace(r'^\d+|\d+$', '')
# + Here we Removes any numbers from the beging and end of every company name
unclean_data_set['Company Name'].tail(1)
```

as we can see that there are still some company names that contain numbers and special caricters

### Removing Rows with false `'Company Names'` - (That Does Not Start with Letters)

```python
for index, row in unclean_data_set.iterrows():
    if not row['Company Name'][0].isalpha():
        # Drop the row if the company name doesn't start with an alphabet
        unclean_data_set = unclean_data_set.drop(index)

unclean_data_set = unclean_data_set.reset_index(drop=True)
unclean_data_set['Company Name'].tail(1)
```

### Turning This `'Salary Estimate'` Unsable Format into another

##### Turning it into (min - max) format

```python
unclean_data_set['Salary Estimate'].value_counts()
```

as we can see, here values are present in 2 formats:

79K−
 131K (Glassdoor est.)
145K−
 225K(Employer est.)
To convert from this format to 79-131 , 145-225

##### Calculate `min_salary`,`max_salary`,`avg_salary`
so that we can use thim in later stages 

```python
unclean_data_set['min_salary'] = 0
unclean_data_set['max_salary'] = 0
unclean_data_set['avg_salary'] = 0
# + we created 3 new series

for i in range(len(unclean_data_set)):
    try:
        # ! From format one - 79K−131K (Glassdoor est.)
        min_salary = int(unclean_data_set['Salary Estimate'][i].split(" ")[0].split("-")[0].replace("$", "").replace("K", ""))
        # + getting the min salary
        max_salary = int(unclean_data_set['Salary Estimate'][i].split(" ")[0].split("-")[1].replace("$", "").replace("K", ""))
        # + getting the max salary
    except:
        # ! From format two - 145K−225K(Employer est.)
        min_salary = int(unclean_data_set['Salary Estimate'][i].split("(E")[0].split("-")[0].replace("$", "").replace("K", ""))
        # + getting the min salary
        max_salary = int(unclean_data_set['Salary Estimate'][i].split("(E")[0].split("-")[1].replace("$", "").replace("K", ""))
        # + getting the max salary
    finally:
        # + assigning the values to there respective cells
        unclean_data_set.loc[i, "min_salary"] = min_salary
        unclean_data_set.loc[i, "max_salary"] = max_salary
        # + reformating the the Salary Estimate
        unclean_data_set.loc[i, "Salary Estimate"] = str(min_salary) + "-" + str(max_salary)
        # + getting the average salary and assigning it to the respective cell
        unclean_data_set.loc[i, "avg_salary"] = (unclean_data_set.loc[i, "min_salary"] + unclean_data_set.loc[i, "max_salary"]) / 2
```

### Formating the `job titles` by simplyfing them and extracting seniority

```python
unclean_data_set['Job Title'].value_counts()[:70]

# +  Displays Common 70
```

```python
def seniority(job_title):
    job_title=job_title.lower()
    snr=['sr','senior','lead','principal','vp','vice president','director']
    for i in snr:
        if i in job_title:
            return "senior"
    if "jr" in job_title:
        return "junior"
    return "na"
```

```python
unclean_data_set['Seniority']=unclean_data_set['Job Title'].apply(seniority)
# + Here we extract the senrority title from the job title and store it in an new series
```

```python
unclean_data_set[unclean_data_set['Job Title']=='Vice President, Biometrics and Clinical Data Management']
# !NOTE
# - as we can see some job titles are too long so we need to simplify them
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    elif 'vice president' in title.lower():
        return 'vp'
    else:
        return 'na'
    
unclean_data_set['job_simp']=unclean_data_set['Job Title'].apply(title_simplifier)

# + here we create a new column with the job_title is Simplefied with out effecting the orignal 
```
